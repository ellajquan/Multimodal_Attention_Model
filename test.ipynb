```python
import boto3
import pandas as pd
from io import BytesIO

# Initialize S3 client
s3 = boto3.client('s3')
bucket_name = 'ella-dlbiomarkers'
file_key = 'CMMD-D2/modified_organized_metadata.xlsx'  # Key is the path within the bucket

# Read file directly from S3 into memory
obj = s3.get_object(Bucket=bucket_name, Key=file_key)
data = pd.read_excel(BytesIO(obj['Body'].read()))  # Load content into BytesIO for compatibility

print(data.head())
```
